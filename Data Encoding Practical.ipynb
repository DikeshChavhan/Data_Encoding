```python
# Nominal / OHE  ---->> binary vector for each of the categaries
# example single, married, seprated
# single: [1,0,0]
# married: [0,1,0]
# seprated: [0,0,1]
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({"status": ["single", "married", "seprated", "single", "married"]})
df
```

```python
encoder = OneHotEncoder()
```

```python
encoder
```

```python
encoded = encoder.fit_transform(df[["status"]]).toarray()
```

```python
encoded
```

```python
encoder.get_feature_names_out()
```

```python
encoder_df = pd.DataFrame(encoded,columns = encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
encoder.transform([["single"]]).toarray()
```

```python
# pd.get_dumies()
```

```python
pd.concat([df, encoder_df], axis= 1)
```

```python
# Target guided ordinal encoding

# relationship with target variable
# when a lot of unique categories
# replace category with mean and median of respective group
```

```python
df = pd.DataFrame({'time':['lunch', 'breakfast', 'dinner', 'lunch', 'breakfast', 'dinner', 'lunch', 'breakfast', 'dinner'],
             'total_bill': [120, 130, 90, 125, 150, 190, 160, 180, 189]})
```

```python
df
```

```python
mean_price = df.groupby("time")["total_bill"].mean().to_dict()
```

```python
df["time_encoded"] = df["time"].map(mean_price)
```

```python
df["time_encoded"]
```
